```python
imagen = cv2.imread('monedas.jpg')
gray = cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
gaussiana = cv2.GaussianBlur(gray, (5, 5), 0)
gaussiana = cv2.GaussianBlur(gaussiana, (5, 5), 0)
gaussiana = cv2.GaussianBlur(gaussiana, (5, 5), 0)


_,th = cv2.threshold(gaussiana,50,gray.max(),cv2.THRESH_BINARY)

contornos1,hierarchy1 = cv2.findContours(th, cv2.RETR_EXTERNAL,
            cv2.CHAIN_APPROX_NONE)
contornos2,hierarchy2 = cv2.findContours(th, cv2.RETR_CCOMP,
            cv2.CHAIN_APPROX_SIMPLE)
```
