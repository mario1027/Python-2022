```python
def cuartiles(dataf,key):
    cuartilesdatos=pd.DataFrame()
    qs = dataf[key].quantile([.25, .50, .75]).values
    cuartilesdatos["Q1"] = (dataf[key] <=qs[0])+0
    cuartilesdatos["Q2"] = ((dataf[key] <=qs[1]) & (dataf[key] >qs[0])) +0
    cuartilesdatos["Q3"] = ((dataf[key] <=qs[2]) & (dataf[key] >qs[1])) +0
    cuartilesdatos["Q4"] = (dataf[key] >qs[2]) +0
    return cuartilesdatos
```

```python
cuartilesdatosmult=datos6.copy()
for i in datos6.drop(columns=["utc_timestamp"]).keys().to_numpy():

    quat=[]
    print(i)
    fff=cuartiles(datos6,i)
    
    for j in datos6.index.to_numpy():
        if fff["Q1"][j]==1:
            quat.append("Q1")
        if fff["Q2"][j]==1:
            quat.append("Q2") 
        if fff["Q3"][j]==1:
            quat.append("Q3")
        if fff["Q4"][j]==1:
            quat.append("Q4")
    
    cuartilesdatosmult[i]=quat
        
        

        
    
        
        
```

```python

fig, ax=plt.subplots(figsize=(16,9))
ax.scatter(valuex,valuey)
ax.boxplot([datos6[cuartilesdatosmult["AT_temperature"]=="Q1"]["AT_radiation_direct_horizontal"].to_numpy(),
            datos6[cuartilesdatosmult["AT_temperature"]=="Q2"]["AT_radiation_direct_horizontal"].to_numpy(),
            datos6[cuartilesdatosmult["AT_temperature"]=="Q3"]["AT_radiation_direct_horizontal"].to_numpy(),
            datos6[cuartilesdatosmult["AT_temperature"]=="Q4"]["AT_radiation_direct_horizontal"].to_numpy()],
            positions=valuex)
```

```python
x = valuex

y = valuey
```

```python
lm_fit = sm.OLS(np.log(y), sm.add_constant(x)).fit()
dt = lm_fit.get_prediction(sm.add_constant(x)).summary_frame(alpha = 0.05)
log_y_prd = dt['mean']
log_y_prd_lower = dt['obs_ci_lower']
log_y_prd_upper = dt['obs_ci_upper']
```
